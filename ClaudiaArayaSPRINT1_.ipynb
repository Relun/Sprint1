# Sprint1 INVENTARIO


```python
#CLASE PRODUCTO
#Se utiliza para crear la clase Producto
#Las clases comienzan con Mayúscula
#Luego se utiliza el constructor init para definir atributos de clase Producto,
#usando self, el cual se utiliza para acceder a los atributos y métodos de esa instancia específica
class Producto:
  def __init__(self, id, nombre, descripcion, cantidad, precio):
    self.id = id
    self.nombre = nombre
    self.descripcion= descripcion
    self.cantidad = cantidad
    self.precio = precio

#Se define el método obtener_informacion que devuelve la información con los atributos de la instancia.
#Se utiliza return para retornar los datos del producto
  def obtener_informacion(self):
    return{
      "id": self.id,
      "nombre": self.nombre,
      "descripcion": self.descripcion,
      "cantidad": self.cantidad,
      "precio": self.precio
      }

#Métodos que permiten actualizan o modificar los atributos correspondientes del producto

#se actualiza solo el atributo nombre
  def actualizar_nombre(self, nombre):
    self.nombre = nombre

#se actualiza solo el atributo descripción
  def actualizar_descripcion(self, descripcion):
    self.descripcion = descripcion

#se actualiza solo el atributo cantidad
  def actualizar_cantidad(self, cantidad):
    self.cantidad = cantidad

#se actualiza solo el atributo precio
  def actualizar_precio(self, precio):
    self.precio = precio

##CLASE INVENTARIO##
#Clase Inventario, permite guardar la información obtenida.
#Se utiliza el constructor init para definir la clase Inventario, se utiliza para manipular las instancias de producto.
#Self se utiliza para acceder a los atributos y métodos de esa instancia específica.
#Se declara el diccionario "productos" como vacío, para ir llenándola con los datos ingresados por el usuario.
class Inventario:
  def __init__(self):
    self.productos = {}

#Método que agrega permite agregar productos a la clase Productos
#Verifica si el id del producto existe y un envía mensaje, en su defecto lo agrega.
#La f en print significa que se puede incorporar una variable en el mensaje de salida.
  def agregar_producto (self, producto):
    if producto.id in self.productos:
      print(f"Producto con id {producto.id} ya existe.")
    else:
      self.productos[producto.id] = producto
      print(f"Producto con id {producto.id} agregado al inventario.")

#Método que elimina el producto de la clase Producto
#Verifica si el id del producto está en lista, lo elimina, en su defecto no lo encuentra y envia un mensaje de salida
  def eliminar_producto(self, id):
    if id in self.productos:
      del self.productos[id]
      print(f"Producto con id {id} eliminado del inventario.")
    else:
      print(f"Producto con id {id} no encontrado en el inventario.")

#Método que actualiza el objeto producto seleccionado por id.
#Verifica si el id del producto está en el diccionario y lo utiliza para un eventual cambio en el valor de algun atributo
#Se establece que los atributos nombre, descripción, cantidad y precio se declaran vacíos por defecto.
#En caso que se ingrese algún valor, dejará de ser vacío y se procederá a cambiar dicho valor.

  def actualizar_producto(self, id, nombre=None, descripcion=None, cantidad=None, precio=None):
    if id in self.productos:
        producto= self.productos[id]
    if nombre is not None:
      producto.actualizar_nombre(nombre)
    if descripcion is not None:
      producto.actualizar_descripcion(descripcion)
    if cantidad is not None:
      producto.actualizar_cantidad(cantidad)
    if precio is not None:
      producto.actualizar_precio(precio)
      #al final del método, alertará el mensaje de que se actualizó el producto del inventario,
      #o en su defecto, no se encontró el id del producto en el inventario
      print(f"Producto con id {id} actualizado en el inventario.")
    else:
      print(f"Producto con id {id} no encontrado en el inventario.")

#Método que obtiene el objeto producto con sus atributos por medio del id.
#Busca el id consultado en el diccionario de productos y si lo encuentra retorna el registro
#consultado a través del método obtener_informacion

  def obtener_informacion_producto(self, id):
    if id in self.productos:
      return self.productos[id].obtener_informacion()
    else:
      return f"Producto con id {id} no encontrado en el inventario."

#Método que recorre el diccionario de todos los productos del inventario, obteniendo y desplegando
#los datos (clave y valor) del diccionario de cada producto.
  def generar_informe(self):
    for producto in self.productos.values():
      print(producto.obtener_informacion())

```

# Funciones adicionales para la interacción con el usuario y manejo de errores

```python
#Función que permite al usuario seleccionar las acciones a realizar
#del sistema de Gestión de Inventario

def mostrar_menu():
  print("\nSistema de Gestión de Inventario")
  print("1. Agregar producto")
  print("2. Eliminar producto")
  print("3. Actualizar producto")
  print("4. Mostrar información de un producto")
  print("5. Generar informe del inventario")
  print("6. Salir")
  return input("Seleccione una opción: ")

#Función reutilizable que valida si es entero o no el valor
#La función valida la excepción de error en el caso que no se ingrese el valor esperado
#según su tipo de dato (en este caso un entero)
def obtener_entero(mensaje):
  while True:
    try:
      return int(input(mensaje))
    except ValueError:
      print("Por favor, ingrese un número entero válido.")

#Función reutilizable que valida si es flotante o no el valor
#La función valida la excepción de error en el caso que no se ingrese el valor esperado
#según su tipo de dato (en este caso un tipo flotante)
def obtener_flotante(mensaje):
  while True:
    try:
      return float(input(mensaje))
    except ValueError:
      print("Por favor, ingrese un número flotante válido.")

```

##Implementación del bucle principal para interactuar con el usuario

```python
#La siguiente función main definirá las acciones a realizar del sistema, permitiendo el llamado de los distintos métodos definidos anteriormente
def main():
  #en primer lugar se instancia la clase Inventario a través del objeto inventario
  inventario =Inventario()
  #Aquí se define las opciones que el usuario podrá seleccionar dentro del menú
  #mientras sea verdadero, el parámetro ingresado mostrará la funcion mostrar_menu
  while True:
    opcion = mostrar_menu()
    #si la opción es igual a 1, entonces invocará el método agregar_producto al inventario
    #también se invocan las funciones reutilizables de obtener_entero y obtener_flotante
    if opcion=='1':
      id = obtener_entero("ID del producto: ")
      nombre= input("Nombre del producto: ")
      descripcion = input("Descripción del producto: ")
      cantidad = obtener_entero("Cantidad: ")
      precio =obtener_flotante("Precio: ")
      producto =Producto(id, nombre, descripcion, cantidad, precio)
      inventario.agregar_producto(producto)
    #si la opción es igual 2, entonces invocará el método eliminar_producto del inventario
    elif opcion == "2":
      id = obtener_entero("ID del producto a eliminar: ")
      inventario.eliminar_producto(id)
    #si la opción es igual 3, entonces invocará el método actualizar_producto del inventario
    elif opcion == '3':
      id = obtener_entero("ID del producto a actualizar: ")
      nombre = input("Nuevo nombre (dejar en blanco para no cambiar): ")
      descripcion = input("Nueva descripción (dejar en blanco para no cambiar): ")
      cantidad_str = input("Nueva cantidad (dejar en blanco para no cambiar): ")
      precio_str=input("Nuevo precio (dejar en blanco para no cambiar): ")
      cantidad = int(cantidad_str) if cantidad_str.isdigit() else None #verifica si la cadena de texto está compuesta de digitos o no, devuelve true/false
      precio = float(precio_str) if precio_str.replace('.','', 1).isdigit() else None  #si viene un punto, este será reemplazado por un caracter vacío
      #si el dato viene vacío lo ingresa como none o en su defecto ingresa el valor
      inventario.actualizar_producto(id, nombre if nombre else None, descripcion if descripcion else None, cantidad if cantidad is not None else None, precio if precio is not None else None)
    #si la opción es igual 4, entonces invocará el método obtener_informacion_produco por id dentro del inventario.
    #si lo encuentra, despliega información y sino imprime que el id no fue encontrado
    elif opcion == '4':
      id=obtener_entero("ID del producto a mostrar: ")
      info=inventario.obtener_informacion_producto(id)
      if info:
        print(info)
      else:
        print(f"Producto con id {id} no encontrado.")
  #si la opción es igual 5, entonces invocará el método generar_informe el cual desplegará todos los productos del inventario
    elif opcion == '5':
        inventario.generar_informe()
#si la opción es igual 6, entonces sale del sistema
#el break funciona para salir del bucle while
    elif opcion == '6':
      print("Saliendo del sistema..")
      break
#si ingresa un valor distinto del 1 al 6, entoces imprimirá como una opción no válida
    else:
      print("Opción no válida. Intente de nuevo.")
 #sentencia que se usa por defecto para invocar al main()
 #el main es para llamar los métodos a otros archivos,
 #permite que se puedan llamar de a uno y en el mismo script en el qué está escrito, ejecuta todo el código
if __name__=="__main__":
  main()
```
